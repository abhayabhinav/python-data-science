# sci kit - Predict flight arrival time - Machine Learning Using Linear Regression model

# Reading flight data and cleaning

```python
delays_df = pd.read_csv('data/Lots_of_flight_data.csv', on_bad_lines='skip')
```

```python
delays_df.info
```

```python
delays_df.info()
```

```python
delays_df.shape
```

```python
delays_cleaned_df = delays_df.dropna()
```

```python
delays_cleaned_df.info()
```

```python
X = delays_cleaned_df.loc[:,['DISTANCE', 'CRS_ELAPSED_TIME']]
```

```python
y = delays_cleaned_df.loc[:,['ARR_DELAY']]
```

# Using scikit-learn to import train_test_split for the prediction

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train, X_test, y_train, y_test
```

```python
X_train.head(), y_train.head()
```

```python
X_test.head(), y_test.head()
```

# Import LinearRegression and use the fit method

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

# Use predict method to check how well data is trained

```python
y_predicted = regressor.predict(X_test)
```

```python
y_predicted , y_test
```

# Check accuracy of the model predictions - Use MSE ( Mean Squared Error )

@ MSE = mean((actual - predicted) ** 2)

```python
from sklearn import metrics
```

```python
mse = metrics.mean_squared_error(y_test,y_predicted)
```

```python
print(f'MSE of the model is {mse}')
```

```python
import numpy as np
```

```python
RMSE = np.sqrt(mse)
```

```python
print(f'Root MSE of the model is {RMSE}')
```
